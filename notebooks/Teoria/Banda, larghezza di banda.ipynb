# Banda e larghezza di banda
Definiamo solo per questo notebook la quantità $M_f$.
$$
M_f = \begin{cases} |c_k| & \mbox{se } x(n) \mbox{ è periodico} \\ 
|X(\omega)| & \mbox{se } x(n) \mbox{ è non periodico} \end{cases} 
$$
$M_f$ è in altre parole, lo spettro di ampiezza del segnale $x(n)$. Da notare che abbiamo parametrizzato $M_f$ con la frequenza $f$, che ricodiamo, è sempre compresanel range  $-\frac{1}{2} \leq f \leq \frac{1}{2}$. Questo perchè sia il parametri $k$ (nel caso di segnali periodici), che il 
parametro $\omega$ sono riferimenti alle frequenze del segnale. 
Volendo eliminare ogni arbitrarietà che si introduce in questo genere di definizioni, diremo che l'intervallo di frequenze su cui lo spettro di ampiezza $M_f$ è non nulla viene chiamato <b> banda </b> ed è dato dall'insieme $\mathcal{F}_{x}=\{f \in [-\frac{1}{2},\frac{1}{2}]\subset \mathbb{R}:M_f \neq 0\} ;$ la sua misura $B_{F}$ è detta invece <b> larghezza di banda </b> (dall'inglese bandwidth). La banda è dunque l'insieme di tutte le frequenze di cui è composto il segnale, mentre la larghezza di banda è lo scarto tra la frequenza più grande e quella più piccola presente nel segnale. Ogni tanto la larghezza di banda si definisce anche associata ad una percentuale. Supponiamo che un segnale abbia il 95% della sua power (o energy) density spectrum all'interno delle frequenze [$f_1$, $f_2$]. Questo vuol dire che per valori esterni a tale intervallo il segnale è molto debole in ampiezza. Perciò, si dice che la larghezza di banda al 95% è concentrata nel range di frequenze $f_1 \leq f \leq f_2$, per cui la larghezza di banda al 95% è pari a $B_{F} = f_2 - f_1$. 

Inoltre, i segnali possono essere definiti sono poi definiti in due modi:
- <b> banda stretta: </b> Se la larghezza di banda $B_{F} = f_2 - f_1$ è molto più piccola (diciamo almeno $\frac{1}{10}$ o ancora meno) della frequenza mediana $\frac{f_2 + f_1}{2}$.
- <b> banda larga: </b> Se non è a banda stretta.
- <b> banda limitata: </b> Un segnale $x(n)$ è detto a banda limitata dalla frequenza $B_F$ se $\mathcal{F}_{x}=(-B_F, B_F)$. 

ATTENZIONE! In realtà, praticamente nessun segnale nella realtà è a banda limitata, ovvero nessun segnale ha range di frequenze al di fuori dei quali la trasformata di fourier è completamente nulla. Perciò di solito ci sono diverse definizioni di bande, che comunque si possono riassumere come l'intervallo di valori di frequenze tali per cui la ESD (Energy Density Spectrum) o PSD (Power Density Spectrum) è apprezzabilmente diversa da zero. Che vuol dire nello specifico apprezzabilmente diversa da zero? Questo dipende a seconda della definizione, qui di seguito ne citiamo alcune.
- Banda a 3 dB: intervallo di valori di frequenze tali per cui l'energia/potenza del segnale si è ridotto della metà.
- Banda al 90%: intervallo di valori di frequenze in cui si concentra il 90% dell'energia/potenza del segnale
- Banda al 95%: intervallo di valori di frequenze in cui si concentra il 95% dell'energia/potenza del segnale

### Banda e larghezza di banda per segnali reali
Per i segnali $x(n)$ reali, abbiamo visto che lo spettro di ampiezza è una funzione pari. Ovvero:
- $|X(-\omega)|=|X(\omega)|$ (simmetria sullo spettro di ampiezza di un segnale $x(n)$ non periodico)
- $\left|c_{-k}\right|=\left|c_{k}\right| \quad$ (simmetria sullo spettro di ampiezza di un segnale $x(n)$ non periodico)

Per questo motivo, è sufficiente considerare solo le frequenze positive, dimezzando di fatto la banda cosiddetta bilatera e riferendoci alla banda monolatera in cui $f \in[0,\frac{1}{2}] \rightarrow \omega \in[0,+\pi] .$ Risultano inoltre di particolare interesse segnali a banda limitata: un segnale $x(n)$ è detto a banda limitata dalla frequenza $B$ se $\mathcal{F}_{x}=(-B, B)$. Distinguiamo nella definizione il caso di segnale discreto periodico e non periodico.
- Un segnale $x(n)$ periodico di periodo N, è a banda limitata se $c_k = 0 \text{ per } k_0 < |k| < N$. Dove $\frac{k_0}{N} = B$. 
- Un segnale $x(n)$ non periodico, è a banda limitata se $|X(\omega)| = 0, \text{ per } \omega_0 < |\omega| < \pi$. Dove $\frac{\omega_0}{2 \pi} = B$.

#### Esempio: segnale periodico a banda non limitata
Consideriamo il segnale nell'esempio del Notebook 03.
$$
x(n)=\begin{cases} A \quad \mbox{ se } \mbox{ n} \in [0+mN, L-1+mN] \\ 0 \quad \mbox{ altrimenti} \end{cases} \quad \text{con }m \in \mathbb{Z}
$$

```python
def f(n, N, A, L):
    m1 = int(n/N)
    m2 = int((n + 1)/N)
    m3 = int((n - 1)/N)
    l1 = np.arange(m1*N,L-1 + m1*N + 1)
    l2 = np.arange(m2*N,L-1 + m2*N + 1)
    l3 = np.arange(m3*N,L-1 + m3*N + 1)
    if n in l1:
        out = A
    elif n in l2:
        out = A
    elif n in l3:
        out = A
    else:
        out = 0
    return out
def calc_x_n(n, N, A, L):
    return np.vectorize(f)(n, N, A, L)
A = 2
N = 9
L = 5
c_k = []
c_0 = A*L/N
c_k.append(c_0)
for k in range(1, N):
    res = (A/N)*(np.cos(k*np.pi*(L-1)/N) - 1j*np.sin(k*np.pi*(L-1)/N))*((np.sin(np.pi*k*L/N))/(np.sin(np.pi*k/N)))
    c_k.append(res)
    print('k: {} \t result: {}'.format(k, np.round(res, 2)))
    
n = np.arange(0.0, 20).astype(int)
x_n = calc_x_n(n, N, A, L)
plt.plot(n, x_n)
plt.title('Segnale periodico')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.xticks(n)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.legend(loc='upper right')
plt.show()

f_m5 = -5/N
f_m4 = -4/N
f_m3 = -3/N
f_m2 = -2/N
f_m1 = -1/N
freqs = [f_m5, f_m4, f_m3,f_m2,f_m1]
for i in range(int((N-1)/2)+2):
    freqs.append(i/N)
c_km5 = c_k[-5]
c_km4 = c_k[-4]
c_km3 = c_k[-3]
c_km2 = c_k[-2]
c_km1 = c_k[-1]
c_k_ = [c_km5, c_km4, c_km3, c_km2, c_km1]
for i in range(int((N-1)/2)+2):
    c_k_.append(c_k[i])

plt.plot(freqs, [abs(i) for i in c_k_])
plt.axvline(x=-0.5, color='r', ls='--')
plt.axvline(x=0.5, color='r', ls='--')
plt.title('Spettro di ampiezza')
plt.xlabel('Frequenza (espressa in cicli per unità di tempo)')
plt.ylabel('$|c_k|$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()

plt.plot(freqs, [abs(i)**2 for i in c_k_])
plt.axvline(x=-0.5, color='r', ls='--')
plt.axvline(x=0.5, color='r', ls='--')
plt.title('Power Density Spectrum')
plt.xlabel('Frequenza (espressa in cicli per unità di tempo)')
plt.ylabel('$|c_k|^2$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()

plt.plot(freqs, [cmath.phase(i) for i in c_k_])
plt.axvline(x=-0.5, color='r', ls='--')
plt.axvline(x=0.5, color='r', ls='--')
plt.title('Spettro di fase')
plt.xlabel('Frequenza (espressa in cicli per unità di tempo)')
plt.ylabel('$\measuredangle c_k$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

Cerchiamo ora di determinare la larghezza di banda. Il segnale non è a banda limitata perchè vediamo che nel range $\mathcal{F}_{x}=\{f \in [0,\frac{1}{2}]\subset \mathbb{R} \}$ lo spettro di ampiezza $|c_k|$ è non nullo $\forall k = 0, 1, ..., N-1$. Tuttavia, la maggior parte della potenza del sengale, si trova nel range $\mathcal{F}_{x}=\{f \in [0,\frac{2}{9}] = [0,0.22] \subset \mathbb{R} \}$. Andiamo a vedere a che percentuale di banda corrisponde tale range di frequenze.

Determiniamo la potenza del segnale. Essendo reale, consideriamo solo la parte positiva dello spettro. La potenza è dunque:
$$
P_{x} =\sum_{k=0}^{\frac{N-1}{2}}\left|c_{k}\right|^{2} = |c_{0}|^{2} + |c_{1}|^{2} + |c_{2}|^{2} + |c_{3}|^{2}+ |c_{4}|^{2}
$$

```python
freqs_positive = freqs[5:10]
c_k_positive = c_k_[5:10]
tot_power = sum([abs(i)**2 for i in c_k_positive])
```

```python
print("La potenza totale nel range f ∈ [0, 0.5] è pari a {}".format(np.round(tot_power, 2)))
```

Consideriamo solo le prime due componenti del PSD, ovvero quelle con frequenza pari a $\frac{0}{N} = \frac{0}{9} = 0, \frac{1}{N} = \frac{1}{9} = 0.11$. 
$$
|c_{1}|^2 = 1.23, \quad |c_{2}|^2 = 0.49, 
$$
Sommando tali componenti, si ottiene:

$$
|c_{1}|^2 + |c_{2}|^2 = 1.23 + 0.49 = 1.64 
$$

Rapportiamo ora 1.64 sulla potenza totale del segnale, e otteniamo:
$$
\frac{|c_{1}|^2 + |c_{2}|^2}{P_x} = \frac{1.64}{1.73} = 95\%
$$

Perciò, considerando la definizione di banda al 95%, possiamo definire $B_F = \frac{2}{9}$, in quanto il 95% della potenza del segnale è racchiuso nell'intervallo $[0, \frac{2}{9}]$, oppure in modo equivalente, nell'intervallo $[-\frac{2}{9}, \frac{2}{9}]$ (essendo il segnale reale). Perciò possiamo scrivere $\mathcal{F}_{x}=(-B_F, B_F) = (-\frac{2}{9}, \frac{2}{9}) $.

```python
plt.plot(freqs, [abs(i)**2 for i in c_k_])
plt.axvline(x=-2/9, color='g', ls='--')
plt.axvline(x=2/9, color='g', ls='--')
plt.axvline(x=-0.5, color='r', ls='--')
plt.axvline(x=0.5, color='r', ls='--')
plt.title('Power Density Spectrum')
plt.xlabel('Frequenza (espressa in cicli per unità di tempo)')
plt.ylabel('$|c_k|^2$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

Come possiamo vedere il segnale è limitato al range $\mathcal{F}_{x}=\{f \in [0,\frac{2}{6}] = [0, 0.33] \subset \mathbb{R} \}$.

#### Esempio - segnale non periodico a banda non limitata
Tutti gli esempi del notebook 4 sono segnali a banda non limitata.

Fare un esempio con la banda e energia del segnale.

Riprendiamo il segnale nel Notebook 04.
$$
x(n)=a^{n} u(n), \quad-1<a<1
$$
Dove $u(n)$ è lo unit step signal, ovvero un segnale che assume 1 se $n\geq0$, e 0 altrimenti. 
La trasformata di Fourier per il segnale $x(n)$ è la seguente
$$
X(\omega)=\frac{1}{1-a e^{-j \omega}}
$$
L'ESD è il seguente.
$$
S_{xx}(\omega) = |X(\omega)|^2 = \frac{1}{1-2 a \cos \omega+a^{2}}
$$
Consideriamo nell'esempio $\omega = 2 \pi f$, con $f = 0.3$ cicli/unità di tempo, e il parametro $a = 0.5$.

```python
def fn(n, a):
    assert abs(a) <= 1
    if n>=0:
        out = a**n
    else:
        out = 0
    return out
def calc_x_n(n, a):
    return [fn(i, a) for i in n] #np.array([fn(i, a) for i in n]) #np.vectorize(fn)(n, a)
fs = 50
n = np.arange(0.0, 20, 1/fs)
f = 1/3
f_hz = f/n[1]
print('Frequenza = {} cicli/unità_tempo = {} Hz'.format(np.round(f, 2), np.round(f_hz, 2)))
a = 0.5
x_n = calc_x_n(n, a)
plt.plot(n, x_n, 'o')
plt.title('Segnale $x(n)$ con $a = {}$'.format(a))
plt.xlabel('Tempo (secondi)')
plt.ylabel('Ampiezza')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.legend(loc='upper right')
plt.show()

def energy_dens_spec(w, a):
    return 1/(1 - 2*a*np.cos(w) + a**2)
def calc_energy_dens_spec(w,a):    
    return np.vectorize(energy_dens_spec)(w,a)
ws = np.linspace(-np.pi, np.pi, len(x_n))
eds = calc_energy_dens_spec(ws, a)
plt.plot(ws, eds)
plt.axvline(x=-np.pi, color='r', ls='--')
plt.axvline(x=np.pi, color='r', ls='--')
plt.title('Energy Density Spectrum')
plt.xlabel('Pulsazioni in radianti per unità di tempo')
plt.ylabel('$S_{xx}(\omega)$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

Essendo il segnale reale, possiamo limitare i calcoli al range di $\omega \in [0, \pi]$.

```python
ws_positive = ws[int(len(ws)/2):]
eds_positive = eds[int(len(eds)/2):]
tot_energy = np.sum(eds_positive)
print("L'energia totale nel range 𝜔 ∈ [0, π] è pari a {}".format(np.round(tot_energy, 2)))
```

Vediamo quant'è l'energia totale se prendiamo il range $\omega \in [0, 2]$.

```python
energy_2 = np.sum(eds_positive[ws_positive <= 2])
print("Se consideriamo il range 𝜔 ∈ [0, 2], l'energia è pari al {} % del totale".format(np.round(energy_2/tot_energy, 2)))
```

Perciò, considerando la definizione di banda al 87%, possiamo definire $B_F = \frac{\omega_0}{2 \pi} = \frac{2}{2 \pi} = \frac{1}{\pi}$, in quanto il 70% dell'energia del segnale è racchiuso nell'intervallo $f \in [0, \frac{1}{\pi}]$ che equivale a dire $ \omega \in [2, \pi]$. Oppure in modo equivalente, nell'intervallo $f \in [-\frac{1}{\pi}, \frac{1}{\pi}]$ (essendo il segnale reale). Perciò possiamo scrivere $\mathcal{F}_{x}=(-B_F, B_F) = (-\frac{1}{\pi}, \frac{1}{\pi}) $.

```python
plt.plot(ws, eds)
plt.axvline(x=-2, color='g', ls='--')
plt.axvline(x=2, color='g', ls='--')
plt.axvline(x=-np.pi, color='r', ls='--')
plt.axvline(x=np.pi, color='r', ls='--')
plt.title('Energy Density Spectrum')
plt.xlabel('Pulsazioni in radianti per unità di tempo')
plt.ylabel('$S_{xx}(\omega)$')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```
