# Analisi nel dominio delle frequenze di un sistema LTI

### Risposta del sistema LTI ad esponenziali complessi
Abbiamo visto che se S è un sistema LTI, allora la risposta del sistema ad un input $x(n)$ può essere definita attraverso la somma convoluzionale.
$$
y(n) = S[x(n)] =  x(n) \circledast h(n) = \sum_{k = - \infty}^{\infty}x(k)h(n-k) 
$$
Dove $h(n)$ non è altro che la risposta del sistema all'impulso elementare delta, ovvero: $$h(n) = S[\delta(n)]$$
Supponiamo ora che il segnale $x(n)$ in input al sistema S sia un esponenziale complesso.
$$
x(n)=A e^{j \omega n}, \quad-\infty<n<\infty
$$
dove $A$ è l'ampiezza e $\omega = 2 \pi f$ è la pulsazione dell'onda misurata in radianti nell'unità di tempo, con $\omega \in [-\pi, \pi] .$ La risposta al sistema è dunque la seguente
$$
\begin{aligned}
y(n) &=\sum_{k=-\infty}^{\infty} h(k)\left[A e^{j \omega(n-k)}\right] \\
&=A\left[\sum_{k=-\infty}^{\infty} h(k) e^{-j \omega k}\right] e^{j \omega n}
\end{aligned}
$$
Possiamo constatare che la quantità tra parentesi non è altro che la trasformata di Fourier della quantità $h(k)$. Possiamo definire tale quantità come $H(\omega)$.
$$
H(\omega)=\sum_{k=-\infty}^{\infty} h(k) e^{-j \omega k}
$$
Ovviamente la funzione $H(\omega)$ esiste se:
$$
\sum_{n=-\infty}^{\infty}|h(n)|<\infty
$$
Possiamo dunque definire la risposta del sistema S (LTI) all'esponenziale complesso $x(n)$ nel seguente modo:
$$
y(n)=A H(\omega) e^{j \omega n} = x(n)H(\omega)
$$
Possiamo constatare che il risultato in output del sistema è pari all'input (che mantiene la stessa frequenza $\omega$), alterato di una componente moltiplicativa $H(\omega)$.
Possiamo ora esprimere $H(\omega)$ in coordinate polari.
$$
H(\omega)=|H(\omega)| e^{j \Theta(\omega)}
$$
Dove $|H(\omega)|$ è il modulo e di $H(\omega)$ e
$$
\Theta(\omega)=\underline{\alpha} H(\omega)
$$
è la fase. Visto che $H(\omega)$ è la trasformata di Fourier di $\{h(k)\},$ possiamo constatare che $H(\omega)$ è periodica con periodo $2 \pi$. Possiamo inoltre definire $h(k)$ in relazione a $H(\omega)$ attraverso l'integrazione come avevamo visto nel Notebook 04. 
$$
h(k)=\frac{1}{2 \pi} \int_{-\pi}^{\pi} H(\omega) e^{j \omega k} d \omega
$$
Possiamo inoltre scomporre $H(\omega)$ nel seguente modo.
$$
\begin{aligned}
H(\omega) &=\sum_{k=-\infty}^{\infty} h(k) e^{-j \omega k} \\
&=\sum_{k=-\infty}^{\infty} h(k) \cos \omega k-j \sum_{k=-\infty}^{\infty} h(k) \sin \omega k \\
&=H_{R}(\omega)+j H_{I}(\omega) \\
&=\sqrt{H_{R}^{2}(\omega)+H_{I}^{2}(\omega)} e^{j \tan ^{-1}\left[H_{l}(\omega) / H_{R}(\omega)\right]}
\end{aligned}
$$
dove $H_{R}(\omega)$ e$ H_{I}(\omega)$ sono le componenti reali e immaginarie di $H(\omega)$, definite come segue.
$$
\begin{aligned}
H_{R}(\omega) &=\sum_{k=-\infty}^{\infty} h(k) \cos \omega k \\
H_{I}(\omega) &=-\sum_{k=-\infty}^{\infty} h(k) \sin \omega k
\end{aligned}
$$
Possiamo inoltre definire gli spettri di ampiezza e fase di $H(\omega),$ espressi in termini di $H_{R}(\omega)$ e $H_{I}(\omega)$.
$$
\begin{aligned}
|H(\omega)| &=\sqrt{H_{R}^{2}(\omega)+H_{I}^{2}(\omega)} \\
\Theta(\omega) &=\tan ^{-1} \frac{H_{I}(\omega)}{H_{R}(\omega)}
\end{aligned}
$$
Notiamo che $H_{R}(\omega)=H_{R}(-\omega)$ e $H_{I}(\omega)=-H_{I}(-\omega)$. Perciò, se conosciamo $|H(\omega)|$ e $\Theta(\omega)$ per $0 \leq \omega \leq \pi$, allora conosciamo $|H(\omega)|$ e $\Theta(\omega)$ anche per $- \pi \leq \omega \leq 0$.
##### L'importanza di $H(\omega)$ 
Dalle considerazioni fatte fino ad ora, possiamo trarre alcune conclusioni. Conoscere la quantità $H(\omega)$ (o equivalentemente conoscere $|H(\omega)|$ e $\Theta(\omega)$), mi permette di conoscere completamente l'effetto del sistema S (LTI) su un qualsiasi input sinusoidale (complesso o non). Possiamo notare che la quantità $|H(\omega)|$ definisce l'amplificazione (se $|H(\omega)| > 1$) o l'attenuazione (se $|H(\omega)| < 1$) imposta dal sistema S sul segnale $x(n)$ sinusoidale. Viceversa la fase $\Theta(\omega)$ definisce lo shift imposto dal sistema S sul segnale in input. Dunque visto che $H(\omega)$ mi permette di conoscere la risposta del sistema LTI su un segnale sinusoidale, si dice che $H(\omega)$ è la <b> risposta in frequenza </b> del sistema. Allo stesso modo, $|H(\omega)|$ è la <b> risposta in ampiezza</b>, mentre $\Theta(\omega)$ è la <b> risposta in fase</b>.

#### Esempio 
Dato il seguente input $x(n)$:
$$
x(n)=A e^{j n\frac{ \pi}{2}}, \quad-\infty<n<\infty
$$
Determinare l'output del sistema S che ha una risposta all'inpulso pari a:
$$
h(n)=\left(\frac{1}{2}\right)^{n} u(n)
$$
Per ottenere $y(n)$, per prima cosa calcoliamo la trasformata di Fourier del segnale $h(n)$
$$
H(\omega)=\sum_{n=-\infty}^{\infty} h(n) e^{-j \omega n}=\frac{1}{1-\frac{1}{2} e^{-j \omega}}
$$
Nel nostro caso abbiamo $\omega=\frac{\pi}{2}$. Perciò:
$$
H\left(\frac{\pi}{2}\right)=\frac{1}{1+j \frac{1}{2}}=\frac{2}{\sqrt{5}} e^{-j 26.6^{\circ}}
$$
A questo punto possiamo calcolare l'output.
$$
\begin{array}{l}
y(n)=A\left(\frac{2}{\sqrt{5}} e^{-j 26.6^{\circ}}\right) e^{j \pi n / 2} \\
y(n)=\frac{2}{\sqrt{5}} A e^{j\left(\pi n / 2-26.6^{\circ}\right)}, \quad-\infty<n<\infty
\end{array}
$$

```python
def f_x_n(A, w, n):
    return A*(np.cos(w*n) + 1j*np.sin(w*n))
def calc_x_n(A, w, n):
    return np.vectorize(f_x_n)(A, w, n)
```

```python
def f_h_n(n):
    if n>=0:
        return (1/2)**n
    else:
        return 0
def calc_h_n(n):
    return np.vectorize(f_h_n)(n)
```

```python
def fourier_transform(w):
    return 1/(1 - (1/2)*(np.cos(w) - 1j*np.sin(w)))
def calc_fourier_transform(w):
    return np.vectorize(fourier_transform)(w)
```

```python
A = 1.5
w = np.pi/2
fs = 50
n = np.arange(0.0, 20, 1/fs)
f = w/(2*np.pi)
f_hz = f/n[1]
print('Frequenza = {} cicli/unità_tempo = {} Hz'.format(np.round(f, 2), np.round(f_hz, 2)))
x_n = calc_x_n(A, w, n)
plt.plot(n, x_n.real, label = 'real')
plt.plot(n, x_n.imag, 'green', label = 'imag')
plt.title('Segnale x(n)')
plt.xlabel('Tempo (secondi)')
plt.ylabel('Ampiezza')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.legend(loc='upper right')
plt.show()
```

```python
h_n = calc_h_n(n)
plt.plot(n, h_n, color = 'k')
plt.title("Risposta del sistema S all'impulso elementare $\\delta(n)$")
plt.xlabel('Tempo (secondi)')
plt.ylabel('Ampiezza')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.legend(loc='upper right')
plt.show()
```

```python
def plot_ft(ft_, ws_rad_unit_tempo, rad_s = True, letter = 'H'):
    ### xlabel = 'rad_s': unità di misura sono i radianti al secondo (la frequenza sono gli Hertz).
    ### altrimenti l'unità di misura sono i radianti per unità di tempo. 
    if rad_s:
        ws = ws_rad_unit_tempo/n[1]
        ft = ft_/n[1]
        #eds = eds_/n[1]
        xlabel = 'Pulsazioni in radianti al secondo'
        lim_inf = -np.pi/n[1]
        lim_sup = np.pi/n[1]
    else:
        ws = np.linspace(-np.pi, np.pi, len(x_n))
        ft = ft_
        #eds = eds_
        xlabel = 'Pulsazioni in radianti per unità di tempo'
        lim_inf = -np.pi
        lim_sup = np.pi
        
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    axs[0, 0].plot(ws, ft.real)
    axs[0, 0].axvline(x=lim_inf, color='r', ls='--')
    axs[0, 0].axvline(x=lim_sup, color='r', ls='--')
    axs[0, 0].set_title('Trasformata di Fourier (parte reale)')
    axs[0, 0].set_xlabel(xlabel)
    axs[0, 0].set_ylabel('R(${}(\omega)$)'.format(letter))
    axs[0, 0].grid(True, which='both')
    axs[0, 0].axhline(y=0, color='k')

    axs[0, 1].plot(ws, ft.imag)
    axs[0, 1].axvline(x=lim_inf, color='r', ls='--')
    axs[0, 1].axvline(x=lim_sup, color='r', ls='--')
    axs[0, 1].set_title('Trasformata di Fourier (parte immaginaria)')
    axs[0, 1].set_xlabel(xlabel)
    axs[0, 1].set_ylabel('I(${}(\omega)$)'.format(letter))
    axs[0, 1].grid(True, which='both')
    axs[0, 1].axhline(y=0, color='k')

    axs[1, 0].plot(ws, [abs(i) for i in ft])
    axs[1, 0].axvline(x=lim_inf, color='r', ls='--')
    axs[1, 0].axvline(x=lim_sup, color='r', ls='--')
    axs[1, 0].set_title('Spettro di ampiezza')
    axs[1, 0].set_xlabel(xlabel)
    axs[1, 0].set_ylabel('|${}(\omega)$|'.format(letter))
    axs[1, 0].grid(True, which='both')
    axs[1, 0].axhline(y=0, color='k')

    axs[1, 1].plot(ws, [cmath.phase(i) for i in ft])
    axs[1, 1].axvline(x=lim_inf, color='r', ls='--')
    axs[1, 1].axvline(x=lim_sup, color='r', ls='--')
    axs[1, 1].set_title('Spettro di fase')
    axs[1, 1].set_xlabel(xlabel)
    axs[1, 1].set_ylabel('$\measuredangle {}(\omega)$'.format(letter))
    axs[1, 1].grid(True, which='both')
    axs[1, 1].axhline(y=0, color='k')
    plt.show()
```

```python
ws = np.linspace(-np.pi, np.pi, len(x_n))
H_ws = calc_fourier_transform(ws)
plot_ft(H_ws, ws, rad_s = False)
```

```python
H_w = calc_fourier_transform(w)
print("Per w = {}, si ha la trasformata di Fourier che è pari a {}".format(np.round(w,2),H_w))
```

```python
print("La risposta in ampiezza |H(w)| è pari a {}".format(np.round(abs(H_w), 2)))
```

Essendo $|H(w)| < 1$, mi aspetto una diminuzione dell'ampiezza originale.

##### Il caso di sinusoidi in combinazione lineare
Uno dei casi più comuni è quando l'input del sistema consiste in una combinazione lineare di sinusoidi, del tipo:
$$
x(n)=\sum_{i=1}^{L} A_{i} \cos \left(\omega_{i} n+\phi_{i}\right), \quad-\infty<n<\infty
$$
dove $\left\{A_{i}\right\}$ e $\left\{\phi_{i}\right\}$ sono le ampiezze e le fasi delle singole componenti sinusoidali. La risposta del sistema S in questo caso è semplicemente
$$
y(n)=\sum_{i=1}^{L} A_{i}\left|H\left(\omega_{i}\right)\right| \cos \left[\omega_{i} n+\phi_{i}+\Theta\left(\omega_{i}\right)\right]
$$
dove $\left|H\left(\omega_{i}\right)\right|$ e $\Theta\left(\omega_{i}\right)$ sono le ampiezze e le fasi, imposte dal sistema alle singole componenti sinusoidali in input. E' chiaro che, a seconda della frequenza del segnale in input, l'output viene affetto in modo diverso dal sistema. Per esempio, alcune sinusoidi possono essere completamente annullate dal sistema. Supponiamo di avere una sinusoide con frequenza $\omega_k$. Se $H(\omega_k)=0$, la sinusoide in input viene soppressa in output. Possiamo dunque considerare il sistema S come un filtro per sinusoidi (complesse e non) di diversa frequenza. Infatti l'idea che c'è nel design dei filtri è proprio quella di determinare i parametri del sistema S per ottenere una determinato output $H(\omega)$ desiderato.

### Risposta del sistema LTI a segnali non periodici
L'output $y(n)$ di un sistema LTI per un segnale $x(n)$ non periodico, è definito sempre dalla convoluzione del segnale in input per la quantità $h(n)$.
$$
y(n) = S[x(n)] =  x(n) \circledast h(n)
$$
Ma come possiamo determinare $y(n)$? Il modo più semplice è quello di sfruttare il teorema della convoluzione, secondo cui:
$$
y(n)=x(n) \circledast  h(n) \stackrel{F}{\longleftrightarrow} Y(\omega)=X(\omega) H(\omega)
$$
In pratica lo spettro dell'output è pari al prodotto tra lo spettro dell'input e la risposta in frequenza del sistema. Possiamo inoltre esprimere lo spettro di fase e di ampiezza di $y(n)$, che vengono di seguito definiti.
$$
|Y(\omega)| = |H(\omega)||X(\omega)| \\
\measuredangle Y(\omega) = \measuredangle H(\omega) + \measuredangle X(\omega)
$$
Possiamo constatare che l'output di un sistema LTI non può contenere frequenze che non erano presenti nell'input. Al massimo può amplificare le frequenze di input, e può cambiare la fase del segnale $x(n)$. Il segnale $y(n)$ può essere determinato nel seguente modo.
$$
y(n)=\frac{1}{2 \pi} \int_{-\pi}^{\pi} Y(\omega) e^{j \omega n} d \omega
$$
L'energia è invece pari a
$$
\begin{aligned}
E_{y} &=\frac{1}{2 \pi} \int_{-\pi}^{\pi} S_{y y}(\omega) d \omega \\
&=\frac{1}{2 \pi} \int_{-\pi}^{\pi}|H(\omega)|^{2}|X(\omega)|^{2} d \omega
\end{aligned}
$$

#### Esempio 
Dato il seguente input $x(n)$:
$$
x(n)=\left(\frac{1}{4}\right)^{n} u(n)
$$
Determinare l'output del sistema S che ha una risposta all'inpulso pari a:
$$
h(n)=\left(\frac{1}{2}\right)^{n} u(n)
$$
Per ottenere $y(n)$, per prima cosa calcoliamo la trasformata di Fourier del segnale $h(n)$
$$
\begin{aligned}
H(\omega) &=\sum_{n=0}^{\infty}\left(\frac{1}{2}\right)^{n} e^{-j \omega n} \\
&=\frac{1}{1-\frac{1}{2} e^{-j \omega}}
\end{aligned}
$$
In modo del tutto analogo, calcoliamo la trasformata di Fourier del segnale $x(n)$.
$$
X(\omega)=\frac{1}{1-\frac{1}{4} e^{-j \omega}}
$$
Lo spettro del segnale in output è dunque il seguente
$$
\begin{aligned}
Y(\omega) &=H(\omega) X(\omega) =\frac{1}{\left(1-\frac{1}{2} e^{-j \omega}\right)\left(1-\frac{1}{4} e^{-j \omega}\right)}
\end{aligned}
$$

```python
def f_x_n(n):
    if n>=0:
        return (1/4)**n
    else:
        return 0
def calc_x_n(n):
    return np.vectorize(f_x_n)(n)
def fourier_transform_H(w):
    return 1/(1 - (1/2)*(np.cos(w) - 1j*np.sin(w)))
def calc_fourier_transform_H(w):
    return np.vectorize(fourier_transform_H)(w)
def fourier_transform_X(w):
    return 1/(1 - (1/4)*(np.cos(w) - 1j*np.sin(w)))
def calc_fourier_transform_X(w):
    return np.vectorize(fourier_transform_X)(w)
```

### Filtri
Abbiamo visto come un sistema LTI modifica lo spettro del segnale in input $X(\omega)$ sulla base della risposta in frequenza $H(\omega)$ resistuendo in output uno spettro $Y(\omega) = X(\omega)H(\omega)$. In un certo senso, $H(\omega)$ agisce come una funzione di pesi che vanno moltiplicati allo spettro del segnale in input. Per questo motivo i termini "filtro" e "sistema LTI" sono spesso usati come sinonimo. Noi utilizzeremo il termine filtro per descrivere un sistema LTI che viene utilizzato per modellare determinate frequenze. 
##### Filtro ideale
Un sistema che annulla le componenti armoniche in determinati intervalli di frequenza e che si comporta come un sistema senza distorsione sulle frequenze rimanenti è detto filtro ideale. Cosa vuol dire non avere distorsione? Un sistema applica una trasformazione non distorta se il segnale in output è una semplice versione amplificata o shiftata del segnale in input. I filtri ideali sono filtri che hanno un gain costante (di solito unitario) in determinati intervalli di frequenza e hanno zero gain altrove. 

Formalmente, i filtri ideali sono del tipo
$$
H(\omega)=\left\{\begin{array}{ll}
C e^{-j \omega n_{0}}, & \omega_{1}<\omega<\omega_{2} \\
0, & \text { altrimenti }
\end{array}\right.
$$
dove $C$ e $n_{0}$ sono costanti. Il segnale in output ha il seguente spettro.
$$
\begin{aligned}
Y(\omega) &=X(\omega) H(\omega) =C X(\omega) e^{-j \omega n_{0}}, \quad \omega_{1}<\omega<\omega_{2}
\end{aligned}
$$
Applicando la proprietà di Frequency Shifting (vedi notebook 04) delle trasformate di Fourier, otteniamo il seguente output nel dominio del tempo.
$$
y(n)=C x\left(n-n_{0}\right)
$$
Di conseguenza, l'output è una semplice versione ritardata e amplificata del segnale di input (quindi non distorta). Perciò i filtri ideali hanno modulo costante e fase lineare, ovvero:
$$
|H(\omega)| = C \\
\Theta(\omega)=-\omega n_{0}
$$
Tuttavia, i filtri ideali non possono essere realizzati in pratica se vogliamo mantenere la causalità (un sistema è detto causale se l'output al tempo n dipende solo da input passati o dall'input presente, ma non da input futuri). Essendo non realizzabili, l'obiettivo è approssimarli. 
### Lowpass, Highpass, Bandpass Filters
Diciamo semplicemente che i filtri digitali possono essere categorizzati in diversi modi. I filtri passabasso (lowpass) hanno un modulo $|H(\omega)|$ che è pari a G (gain) in corrispondenza dello 0, e il gain tende a zero man mano che si procede verso $\pi$. I filtri passaalto (highpass) hanno invece il massimo valore intorno alla pulsazione $\omega = \pi$ e hanno circa 0 gain intorno a $\omega = 0$. I filtri bandpass, invece hanno un modulo $|H(\omega)|$ che è pari a G (gain) in corrispondenza circa di $\omega = \frac{\pi}{2}$ e hanno circa 0 gain intorno a $\omega = 0$ e $\omega = \pi$.
